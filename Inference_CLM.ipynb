```python
# import bitsandbytes as bnb
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer,BitsAndBytesConfig
# from peft import PeftModel, PeftConfig

# bnb_config = BitsAndBytesConfig(
#     load_in_8bit=True,
#     load_8bit_use_double_quant=True,
# )

# bnb_config = BitsAndBytesConfig(
#     load_in_4bit=True,
#     load_4bit_use_double_quant=True,
#     bnb_4bit_quant_type="nf4",
#     # bnb_4bit_compute_dtype=torch.bfloat16,
# )

# Make sure that torch is configured to use the GPU if available
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# print(f"Using device: {device}")

def generate_and_save_text(model_name_or_path, prompt, output_file):
    # Load the model and tokenizer onto the GPU
    # model = AutoModelForCausalLM.from_pretrained(model_name_or_path,quantization_config=bnb_config)
    model = AutoModelForCausalLM.from_pretrained(model_name_or_path)
    # config = PeftConfig.from_pretrained(lora)
    # model = PeftModel.from_pretrained(model,lora)
    tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)

    # Model-specific configuration for creativity
    model_config = {
        "max_length": 1000,
        "min_length": 1000,
        "num_return_sequences": 1,
        "top_p": 0.9,
        "top_k": 50,
        "temperature": 1.2,
        "num_beams": 5,
        "no_repeat_ngram_size": 2,
        "length_penalty": 0.8,
        "penalty_alpha": 0.8,
    }

    # Create a text generation pipeline
    generator = pipeline('text-generation', model=model, tokenizer=tokenizer,device=0)

    # Generate creative text
    generated_text = generator(
        prompt,
        max_length=model_config["max_length"],
        min_length=model_config["min_length"],
        num_return_sequences=model_config["num_return_sequences"],
        top_p=model_config["top_p"],
        top_k=model_config["top_k"],
        temperature=model_config["temperature"],
        num_beams=model_config["num_beams"],
        no_repeat_ngram_size=model_config["no_repeat_ngram_size"],
        length_penalty=model_config["length_penalty"],
        do_sample=True,
        penalty_alpha=model_config["penalty_alpha"],
        use_cache=True
    )[0]['generated_text']

    # Save the generated text to a text file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(generated_text)

    print(f"Generated text saved to {output_file}")

# Example usage
model_name_or_path = 'faizalnf1800/QuantumQuill-300m-GPT2-Medium-Instruct'
# lora = "faizalnf1800/peft_lora_falcon_1b_chat_scifi_webnovel"
prompt = """
Create the opening scene of a sci-fi story set in a distant galaxy, where an interstellar detective is called to investigate a perplexing anomaly on an uncharted planet.

Prologue :
"""
output_file = 'generated_text.txt'

generate_and_save_text(model_name_or_path, prompt, output_file)
```
